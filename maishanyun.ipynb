```python
# Simple: read files into variables
import pandas as pd
from pathlib import Path

root = Path('.')

# Read CSVs into variables
shipment_path = root / 'MSY Data - Shipment.csv'
shipment = pd.read_csv(shipment_path) if shipment_path.exists() else None

ingredient_path = root / 'MSY Data - Ingredient.csv'
ingredients = pd.read_csv(ingredient_path) if ingredient_path.exists() else None

print('shipment loaded' if shipment is not None else 'shipment not found')
print('ingredients loaded' if ingredients is not None else 'ingredients not found')

# Read all monthly Excel files into a dict and combine
data_dir = Path('monthdata')
monthly = {}
combined_monthly = None
if data_dir.exists() and data_dir.is_dir():
    for f in sorted(data_dir.glob('*.xlsx')):
        key = f.stem.replace(' ', '_').replace('-', '_')
        monthly[key] = pd.read_excel(f)
    if monthly:
        combined_monthly = pd.concat(monthly.values(), ignore_index=True)
        print(f"combined_monthly created with {len(combined_monthly)} rows")
    else:
        print('no monthly excel files found in monthdata')
else:
    print('monthdata directory not found')
```
